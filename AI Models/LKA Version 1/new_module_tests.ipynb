```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from LKA_module import LaneDetector, LaneDetectorResult
```

```python
frame = cv2.imread("Screenshot from 2024-02-21 06-00-37.png")
# Check if the image was loaded successfully
if frame is  None:
    print("Failed to load the image.")
else:
    plt.imshow(frame)
```

```python
lower_color = np.array([0, 0, 0])
upper_color = np.array([30, 255, 255])
roi_parm = (0, 0.5, 1)

detector = LaneDetector(lower_color, upper_color, roi_parm=roi_parm)
```

```python
roi = detector.ROI(frame)



plt.imshow(roi)
```

```python
lines = detector.detect_lanes(frame)

print(lines)
```

```python
lane_detection_result = LaneDetectorResult(frame, lines)
error = lane_detection_result.calculate_error()
error
```

```python
lanes = lane_detection_result.plot_lanes_and_center_line()
plt.imshow(lanes)
```
